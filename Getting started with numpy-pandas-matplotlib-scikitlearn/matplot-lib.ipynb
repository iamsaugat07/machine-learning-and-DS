# Matplotlib example workflow

```python
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize = (10,10))

# 3. Plot data
ax.plot(x,y)

#4. Customize plot
ax.set(title = "Simple-plot",
      xlabel = "x-axis",
      ylabel = "y-axis")

#5. Save the plot
fig.savefig("image.png")
```

##  Plotting from pandas DataFrames

```python
import pandas as pd 
#make a dataframe
car_sales = pd.read_csv("car-sales.csv")
car_sales
```

```python
car_sales["Price"] = car_sales["Price"].str.replace('[\$\,\.]','')
car_sales
```

```python
car_sales["Price"] = car_sales["Price"].str[:-2]
car_sales
```

```python
car_sales["Sale Date"] = pd.date_range("1/1/2023", periods = len(car_sales))
car_sales
```

```python
car_sales["Total sales"] = car_sales["Price"].astype(int).cumsum()
car_sales
```

```python
# lets plot the total sales
car_sales.plot(x="Sale Date", y = "Total sales");
```

```python
car_sales.info()
```

```python
car_sales["Price"] = car_sales["Price"].astype(int)
car_sales.plot(x = "Odometer (KM)", y = "Price", kind = "scatter");
```

## Using matplotlib OO method

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50
```

```python
## OO method mixed with pyplot
fig, ax = plt.subplots(figsize = (10,6))
over_50.plot(kind = "scatter",
            x = 'age',
            y = 'chol',
            c = 'target',
            ax = ax);
ax.set_xlim(45,100);
```

```python
heart_disease.head(5)
```

```python
## OO method from scratch

#create a subplot
fig, ax = plt.subplots(figsize = (10,6))

#plot the data
scatter = ax.scatter(x = over_50["age"],
                    y = over_50["chol"],
                    c = over_50["target"])

#customize the plot
ax.set(title = 'Heart Disease and Cholestrol Levels',
       xlabel = "Age",
      ylabel = "Cholestrol")
# add legends 
ax.legend(*scatter.legend_elements(), title = 'Target')

# add a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle = '--');
```

```python
# subplot of chol, age & thalach
fig, (ax0, ax1) = plt.subplots(nrows = 2,
                               ncols = 1,
                              figsize = (10,10),
                              sharex = True)

#add data to subplot
scatter = ax0.scatter(x = over_50["age"],
                      y = over_50["chol"],
                      c = over_50["target"])

# Customize the plot 
ax0.set(title = "Heart disease and Cholestrol level",
       
        ylabel = 'cholestrol')

# add the legends 
ax0.legend(*scatter.legend_elements(), title = "Target")

# add a horizontal line
ax0.axhline(over_50["chol"].mean(),
          linestyle = '--');

# add second subplot 
scatter = ax1.scatter(x = over_50["age"],
                      y = over_50["thalach"],
                      c = over_50["target"])
ax1.set(title = "Heart disease and Max heart rate",
        xlabel = 'age',
        ylabel = 'Max Heart Rate')

ax1.legend(*scatter.legend_elements(), title = "Target")
ax1.axhline(over_50["thalach"].mean(),
          linestyle = '--');

```
